```python
import math
def calcular_pH(h):
    try:
        pH = -1 * math.log10(h)
        return pH
    except ValueError:
        print("H+ nÃ£o pode ser zero ou negativo.")

volume = 100
h = 0.01

print(f"O pH Ã©: {calcular_pH(h)}")
```

3. Crie uma funÃ§Ã£o que faÃ§a a soma o peso total de uma proteÃ­na de acordo com o peso de cada aminoÃ¡cido. Esta funÃ§Ã£o deve criar um except para caso a proteÃ­na contenha algum caracter ("aminoÃ¡cido") que nÃ£o Ã© chave do dicionÃ¡rio e outro except para caso algum valor do dicionÃ¡rio nÃ£o esteja numÃ©rico


```python
#exemplo 1
def calcular_massa_proteina(proteina, aminos_massas):
    try:
        # Calcula o peso total da proteÃ­na
        peso_total = 0
        for amino_acid in proteina:
          peso_total += aminos_massas[amino_acid]

        return peso_total

    except KeyError:
        print(f"Erro: {amino_acid} Ã© um aminoÃ¡cido invÃ¡lido ."
        "Retire-o da sequÃªncia e tente novamente")

    except TypeError:
        print(f"Erro: {amino_acid} estÃ¡ com valor nÃ£o numÃ©rico"
        " no banco de dados (dicionÃ¡rio).")


aminos_massas = {
      'A': 71.04, 'C': 103.01, 'D': '115.03', 'E': 129.04, 'F': 147.07,
      'G': 57.02, 'H': 137.06, 'I': 113.08, 'K': 128.09, 'L': 113.08,
      'M': 131.04, 'N': 114.04, 'P': 97.05, 'Q': 128.06, 'R': 156.10,
      'S': 87.03, 'T': 101.05, 'V': 99.07, 'W': 186.08, 'Y': 163.06
}

proteina="AAAAA-"
peso_total = calcular_massa_proteina(proteina, aminos_massas)
if peso_total:
  print(peso_total)
```

â­ **DESAFIO:** Implemente uma funÃ§Ã£o em Python que calcule o **modelo exponencial** de declÃ­nio da populaÃ§Ã£o.

O modelo leva em conta apenas a taxa de mortalidade, representada pelo parÃ¢metro `rate`. A funÃ§Ã£o deve receber como parÃ¢metros: `initial_population`, `rate`, `years`.

A funÃ§Ã£o deve retornar a populaÃ§Ã£o final.

```python
class SequenciaCurta(Exception):
    pass

def ler_sequencias_fasta(nome_arquivo):
  try:
    with open(nome_arquivo, 'r') as arquivo:
      sequencias = {}
      id_sequencia = ""
      for linha in arquivo:
        print(linha)
        if linha.startswith(">"):
            id_sequencia = linha[1:].strip()
            sequencias[id_sequencia] = ""
        else:
            sequencia = linha.strip()
            if len(sequencia) < 20:
              raise SequenciaCurta(f"A sequÃªncia '{sequencia}' contÃ©m menos de 20 caracteres)")
              sequencias[id_sequencia] += sequencia
      return sequencias
  except FileNotFoundError:
        print(f"O arquivo '{nome_arquivo}' nÃ£o foi encontrado.")
        return None


nome_arquivo = input()
sequencias = ler_sequencias_fasta(nome_arquivo)

if sequencias:
    print(sequencias)
```
