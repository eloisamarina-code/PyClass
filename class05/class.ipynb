Depois de ler o contedo do arquivo em uma varivel, podemos trat-los como qualquer outra string - por exemplo, podemos imprimi-los:

## **Gravar e fechar arquivos**

Aps abrir um arquivo,  possvel alterar o contedo e aps essa alterao, devemos gravar e fechar o arquivo. Para isso, utilizamos o **mtodo close()**.

Devemos chamar close() depois que terminarmos de ler ou gravar em um arquivo, pois evita alguns tipos de bugs que podem ser difceis de rastrear --> Boa Prtica de Python (Pythonic).

close ()  um mtodo incomum, pois no aceita argumentos (por isso  chamado com um par vazio de parnteses) e no retorna nenhum valor til:

A segunda maneira de fechar um arquivo  usando a **instruo with**.

O comando with automaticamente se encarrega de fechar o arquivo assim que ele sai do bloco with, mesmo em casos de erro.

 recomendado preferir o uso da instruo with, pois permite um **cdigo mais limpo** e facilita o **tratamento de erros inesperados**.

Existe um segundo argumento que  uma string que contm vrios caracteres para representar como voc deseja abrir o arquivo. O padro mais comum  **r**, que representa a abertura do arquivo no modo somente leitura como um arquivo de texto:

```python
# Abre o arquivo usando o mtodo with
with open('/content/drive/MyDrive/2 - Grupos de Estudos/2023 1/1. Python Basico/Trabalhando com arquivos/genoma.txt', 'r') as file:
    # Realiza operaes no arquivo
    for line in file:
        # Exemplo: imprimir cada linha do arquivo
        print(line)
        # se a linha comea com determinado conjunto de caracteres
        # retornar posies especficas
```

Podemos verificar se o arquivo foi fechado automaticamente:

```python
# Testando se o arquivo foi fechado...
file.closed
```

## **Manipulando arquivos no terminal do Linux**

Conforme trabalhamos com Bioinformtica, sentimos a necessidade de melhorar nosso conhecimento sobre sistemas operacionais e como manipular nossos arquivos neles.

Quando falamos em terminal, logo j vem uma imagem de alguma distribuio Linux na cabea, porm podemos utilizar o terminal no Mac tambm, alguns comandos so idnticos entre esses sistemas.

No Windows, tambm temos a possibilidade de utilizar o terminal, nele  chamado de cmd, ou prompt de comando, porm no Linux e no Mac essa prtica  mais comum. Por isso, os comando que veremos pelo terminal podero ser usados no Linux e no Mac, mas no so compatveis com o Windows.

No vamos nos aprofundar sobre a linguagem Bash, utilizada no terminal Linux, mas vamos passar alguns comandos teis, caso precise trabalhar nele.

* Quando abrimos o terminal, na maioria das distribuies Linux, vamos estar na **pasta home** do usurio que estamos utilizando.

* Podemos falar para o terminal mostrar para a gente o diretrio atual, ou seja, imprimir o diretrio que estamos trabalhando (pwd, print working directory): `pwd`

* Podemos falar para o terminal listar (ls, list) o seu contedo: `ls`

* Se queremos mudar de diretrio, basta colocar essa informao em ingls (cd, change directory): `cd` `+ nome do diretrio de interesse` ou se quisermos sair desse diretrio: `cd ..`

* Um comando que  utilizado para criar arquivos no sistema  o `touch`, com ele conseguimos, entre outras coisas, criar arquivos no sistema, exemplo: `estudos.py`

* Para criar um diretrio (mkdir, make directory) onde podemos colocar nossos arquivos de estudos: `mkdir Pyladies`

* Ns podemos mover (`mv`) o nosso arquivo `estudos.py` para dentro do diretrio `Pyladies`, assim no precisamos criar outro arquivo

* Podemos remover um arquivo (rm, remove) um arquivo: `rm coletando_dados.py` e tambm diretrios: `rm Pyladies/`. Quando queremos excluir o diretrio COM todos os arquivos contidos nele, inserimos o -R de (recursive): `rm -R Pyladies/`
